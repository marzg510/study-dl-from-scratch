#### 1.6.1 単純なグラフの描画

#### 1.6.2 pyplotの機能

```python
import numpy as np
import matplotlib.pyplot as plt

# データの作成
x =  np.arange(0, 6, 0.1) # ０から６まで０．１刻みで生成
y1 = np.sin(x)
y2 = np.cos(x)

# グラフの描画
plt.plot(x,y1,label="sin")
plt.plot(x,y2,linestyle="--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()
plt.show()
```
