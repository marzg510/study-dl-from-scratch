### 3.1 パーセプトロンからニューラルネットワークへ

#### 3.1.2 パーセプトロンの復習

```python
# ２入力のパーセプトロン
from graphviz import Digraph

dg = Digraph(format='png')
dg.attr('node', shape='circle')

dg.node('x1')
dg.node('x2')
dg.node('y')
dg.edge('x1', 'y', 'w1')
dg.edge('x2', 'y', 'w2')

dg
```

$$
y=\left\{
\begin{array}{ll}
0 & (b + w_1x_1 + w_2x_2 \leq 0) \\
1 & (b + w_1x_1 + w_2x_2 > 0) \\
\end{array} \right.
$$

```python
# バイアスを明示的に示す
from graphviz import Digraph

dg = Digraph(format='png')
dg.attr('node', shape='circle')
dg.attr('graph', rankdir='LR')

dg.node('1')
dg.node('x1')
dg.node('x2')
dg.node('y')
dg.edge('1', 'y', 'b')
dg.edge('x1', 'y', 'w1')
dg.edge('x2', 'y', 'w2')

dg
```
