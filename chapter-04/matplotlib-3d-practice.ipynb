```python
x = np.arange(-3, 3, 0.25)
y = np.arange(-3, 3, 0.25)
X, Y = np.meshgrid(x, y)
Z = np.sin(X)+ np.cos(Y)
```

```python
Z
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(X,Y,Z)
```

```python
x = np.arange(-3, 3, 0.1)
y = np.arange(-3, 3, 0.1)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
Z = X ** 2 + Y ** 2
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(X,Y,Z)
```
