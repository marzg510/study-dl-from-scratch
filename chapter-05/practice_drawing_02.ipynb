```python
%pylab inline --no-import-all
import matplotlib.patches as patches
```

```python
fig = plt.figure()
ax = plt.axes()

# fc = face color, ec = edge color
c = patches.Circle(xy=(0, 0), radius=10, fc='w', ec='k')
a = patches.Arrow(0,0,15,0, fc='k',ec='k')
ax.add_patch(a)
ax.add_patch(c)

plt.axis('scaled')
ax.set_aspect('equal')

plt.show()
```
